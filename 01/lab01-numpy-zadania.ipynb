## Zadania 1.1

1. Za pomocą jednowierszowego polecenia utwórz następującą macierz:
$a =
\begin{pmatrix}
1       & 2      & \cdots & 10      \\
11      & 12     & \cdots & 20      \\
\vdots  & \ddots & \ddots & \vdots  \\
41      & 42     & \cdots & 50 
\end{pmatrix}$
2. Za pomocą metod obiektu `array`: 
 * określ liczbę elementów, kolumn i wierszy,
 * stwórz wektory średnich po wierszach oraz po kolumnach,
 * wypisz trzecią kolumnę macierzy `a`.
3. Przetestuj i opisz w jednym zdaniu działanie następujących funkcji: 
 * `np.log(a)` 
 * `np.cumsum(a)`
 * `np.rank(a)`
 * `np.power(a,2)`
4. Stwórz wektor składający się z 50 losowych punktów z przedziału 14 do 31

```python
# tutaj wpisz swoje rozwiązania zadań 1.1
import numpy as np

#1
a = np.arange(1,51).reshape(5,10)
print("#1")
print(a)
print("\n")

#2
a_size = a.size
a_rows, a_columns = a.shape
column_mean = a.mean(axis=0) # średnia po kolumnach
row_mean = a.mean(axis=1) # średnia po wierszach
print("#2")
print("elements: %d, rows: %d, columns: %d" % (a_size, a_rows, a_columns))
print("mean of columns: %s; mean of rows: %s" % (str(column_mean), str(row_mean)))
print("third column %s" % str(a[:,2]))
print("\n")

#3
np.log(a) # logarytm naturalny każdego elementu
np.cumsum(a) # cumulative sum, w kaźdej komórce znajduje się suma tego elementu i elementów występujących wcześniej
np.ndim(a) # zwaraca liczbę wymiarów, np.rank zostało wycofane (is deprecated)
np.power(a,2) # każdy element tablicy podnosi do kwadratu

#4
vec = np.random.uniform(14,32,50)
#lub np.random.randint(14,32,50) jeśli chodziło o całkowite, bo trochę polecenie jest dla mnie nieprecyzyjne
print("#4")
print(vec)
```

## Zadania 1.2

Zapisz i oblicz za pomocą NumPy
1. Iloczyn macierzy $A$ z wektorem $\vec{x}$:

    \end{array}
  \right]
  \left[

  \left[
    \begin{array}{r}
      1\\
      -3
    \end{array}
  \right].
\end{align*}$$

2. Iloczyn macierzy $A$ i $B$:  
$$\begin{align*}
  AB &=\left[
      \begin{array}{rrr}
        0 & 4 & -2\\
        -4 & -3 & 0
      \end{array}
    \right] 
    \left[
      \begin{array}{rr}
        0 &1\\
        1 & -1\\
        2 & 3
      \end{array}
    \right]
    =
    \left[
      \begin{array}{rr}
        0 & -10\\
        -3 & -1
      \end{array}
    \right].
  \end{align*}$$
  
3. Pokaż, że dla powyższych macierzy $A$ i $B$ prawdą jest, że $(AB)^T = B^TA^T$.  
4. Oblicz $\det(AB)$ (wyznacznik iloczynu $AB$).
4. Czym różni się operacja `A**-1` dla obiektów typu `array` i `matrix`? Pokaż na przykładzie. 
6. Dla macierzy $X = \left[
      \begin{array}{rrr}
        1 & 2 & 3\\
        1 & 3 & 6 \\
      \end{array}
    \right]$ oraz wektora $\vec{y} = \left[
      \begin{array}{r}
        5 \\
        6 \\
      \end{array}
    \right]$ oblicz wynikowy wektor: 
$$\vec{\theta} = (X^TX)^{-1}X^T\vec{y}.$$ Wykonaj te same obliczenia raz na obiektach typu `array` i raz na obiektach typu `matrix`. W przypadku obiektów typu `matrix` zastosuj możliwie krótki zapis. 

```python
# tutaj wpisz swoje rozwiązania zadań 1.2
import numpy as np

#1
A = np.array([[1,-1,2],[0,-3,1]])
x = np.array([2,1,0])
C = np.dot(A,x)
print("#1")
print(C)
print("\n")

#2
A = np.array([[0,4,-2],[-4,-3,0]])
B = np.array([[0,1],[1,-1],[2,3]])
C = np.dot(A,B)
print("#2")
print(C)
print("\n")

#3
C = np.dot(A,B).T
D = np.dot(B.T,A.T)
is_equal = np.array_equal(C,D)
print("#3")
print(is_equal)
print("\n")

#4
detAB = np.linalg.det(np.dot(A,B))
print("#4")
print(detAB)
print("\n")

#5
A_arr = (np.array([[1,2],[3,4]])).astype(float)
A_matr = np.matrix([[1,2],[3,4]])

print("#5")
print(A_matr**-1)
print(A_arr**-1)
print("\n")

# Różnica jest taka, że dla obiektu array operator ** każdy element kolekcji podnosi do danej potęgi (w tym przypadku -1), 
# dla obiektu matrix natomiast wyznacza macierz odwrotną.

#6
X_arr = np.array([[1,2,3],[1,3,6]]).astype(float)
y_arr = np.array([[5],[6]]).astype(float)
theta_arr = np.dot(np.dot(np.linalg.inv(np.dot(X_arr.T,X_arr)),X_arr.T),y_arr)

X_matr = np.matrix([[1,2,3],[1,3,6]])
y_matr = np.matrix([[5],[6]])
theta_matr = ((X_matr.T*X_matr)**-1)*X_matr.T*y_matr

print("#6")
print(theta_arr)
print(theta_matr)
```
